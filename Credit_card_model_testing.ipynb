```python
test_transaction = [
    123456,  # Time (FIRST)
    -1.23, 0.45, 0.12, -0.87, 0.34,  # V1-V5
    -0.56, 0.78, -0.91, 0.23, -0.45,  # V6-V10
    0.67, -0.12, 0.89, -0.34, 0.56,   # V11-V15
    -0.78, 0.91, -0.23, 0.45, -0.67,  # V16-V20
    0.12, -0.89, 0.34, -0.56, 0.78,   # V21-V25
    -0.91, 0.23, -0.45,  # V26-V28
    250.00  # Amount (LAST)
]

def predict_fraud(transaction_features):
    """Returns fraud probability and prediction"""
    # Convert to numpy array and reshape
    input_data = np.array(transaction_features).reshape(1, -1)

    # Apply preprocessing (if you used any during training)
    # input_data = scaler.transform(input_data)  # Uncomment if you scaled data

    # Predict
    fraud_prob = model.predict(input_data)[0][0]
    is_fraud = fraud_prob > 0.992  # Your optimal threshold

    return fraud_prob, is_fraud

# Test the function
probability, prediction = predict_fraud(test_transaction)
print(f"Fraud Probability: {probability:.4f}")
print("Fraud Detected!" if prediction else "Legitimate Transaction")
```
